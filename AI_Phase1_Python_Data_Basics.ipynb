```python
import pandas as pd
data = {
    "name":["kellis", "jack", "martin"],
    "age": [30,33,45],
    "salary": [450000,5000000,600000]
}

dt = pd.DataFrame(data)
dt
print(dt.head())
print(dt.describe())
#young_people = df[df["age"]]
young_people = dt[dt["age"]<33]
print("young people:")
print(young_people)
highly_paid = dt[dt["salary"]>600000]
print("highly paid")
print(highly_paid)
```
