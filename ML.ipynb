```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error


#Simple dataset
experience = np.array([1,2,3,4,5,6,7,8]).reshape(-1,1)
salary = np.array([10000,20000,30000,40000,50000,60000,70000,80000])

df = pd.DataFrame({"Experience":experience.flatten(), "Salary": salary})
plt.figure()
plt.scatter(df["Experience"],df["Salary"])
plt.xlabel("Experience (Years)")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.show()
X = df[["Experience"]]
y = df["Salary"]
# Features (input)
# Target (output)
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.25, random_state=42
)
#0.25 means 25 percent for testing and 75 percent for training 
print("Training size:", len(X_train))
print("Testing size:", len(X_test))
model = LinearRegression()
model.fit(X_train, y_train)
print("Slope (m):", model.coef_) #magnitude of impact
print("Intercept (b):", model.intercept_) #
m = model.coef_[0]       # slope
b = model.intercept_    # intercept

print("Slope (m):", m)
print("Intercept (b):", b)
Salary = m * experience + b
y_pred = model.predict(X_test)
comparison = pd.DataFrame({
    "Actual Salary": y_test.values,
    "Predicted Salary": y_pred.astype(int)
})

comparison

# ==============================
# 8 Evaluate the model
# ==============================

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
    
print("MAE:", mae)
print("RMSE:", rmse)

# ==============================
# 9 Visualization: Actual vs Predicted
# ==============================

plt.figure()
plt.scatter(X_test, y_test, label="Actual")
plt.plot(X_test, y_pred, label="Predicted")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Actual vs Predicted Salary")
plt.legend()
plt.show()

# ==============================
#  Predict salary for new data
# ==============================

new_experience = np.array([[12]])
predicted_salary = model.predict(new_experience)

print("Predicted salary for 10 years experience:", int(predicted_salary[0]))
```
